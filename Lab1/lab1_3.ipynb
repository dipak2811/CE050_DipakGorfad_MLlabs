```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv('/content/Iris.csv')
data.head()
```

**1**

```python
plt.scatter(data['SepalLengthCm'],data['SepalWidthCm'])
plt.title("Scatter Plot of SepalLength and SepalWidth ")
plt.xlabel("SepalLength in Cm")
plt.ylabel("SepalWidth in Cm")
plt.show()
```

**2**

```python
plt.hist(data['SepalLengthCm'])
plt.title("Histogram of SepalLength")
plt.xlabel("SepalLength in Cm")
plt.ylabel("Frequency")
plt.show()
```

**3**

```python
plt.bar(data['Species'],data['Id'])
plt.xlabel("Species")
plt.ylabel("Id")
plt.show()
```
